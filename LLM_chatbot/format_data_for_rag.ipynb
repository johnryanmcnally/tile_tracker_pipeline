```python
def format_csv(df):
    df_format = pd.DataFrame({'id':[],'text':[]})
    text = []
    id = []
    for i, row in df.iterrows():
        row_text = [f"{col}: {row[col]}" for col in df.columns if 'unnamed' not in col.lower()]
        text.append(', '.join(row_text))
        id.append(str(i))
    df_format['text'] = text
    df_format['id'] = id
    return df_format
```

```python
files = Path(r"C:\Users\joyam\Documents\JohnProjects\tile_project\data\staged").glob("*.csv")
for file in files:
    print(file.stem)
    df = pd.read_csv(file)
    df_format = format_csv(df)
    df_format.to_csv(r"C:\Users\joyam\Documents\JohnProjects\tile_project\data\rag_format\\" + f"{file.stem}.csv")
```
